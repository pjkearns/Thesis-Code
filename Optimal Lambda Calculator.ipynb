```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
```

```python
from FH_Functions import N_o_f
```

```python
err   = .01
N2err = []
D2err = []

lams = np.linspace(0, 1.8, 20)

for lam in lams:
    x, n, d = N_o_f(err,lam)
    N2err.append(float(n))
    D2err.append(d)

N2err = np.array(N2err)
D2err = np.array(D2err)


f1 = plt.figure(figsize=(7,5))
plt.rc('font',family='serif')
plt.tick_params(labelsize=20)
plt.plot(lams, N2err, linewidth=3)
# plt.title("Lambda vs. Tt", fontsize=20)
plt.xlabel(r'$\lambda$', fontsize=20)
plt.ylabel(r'Samples to E$(\mathrm{\varepsilon})$ < 0.01', fontsize=20);

f1.tight_layout()
f1.savefig('N_vs_Lambda.eps', format='eps', dpi=100, bbox_inches='tight');
```

```python
err   = .01
N2err = []
D2err = []

lams = np.linspace(0.01, 1.50, 150)

for lam in lams:
    x, n, d = N_o_f(err,lam)
    N2err.append(float(n))
    D2err.append(d)

N2err = np.array(N2err)
D2err = np.array(D2err)
```

```python
Ts = 100
min_lam = []
nRatio = 1000
ratio = np.logspace(-4,3,nRatio)
Tts = Ts*ratio
for Tt in Tts:
    T = np.sum([Ts*N2err, Tt*D2err], axis=0)
    min_lam.append(lams[np.argmin(T)])
```

```python
f1 = plt.figure(figsize=(7,5))
plt.rc('font',family='serif')
plt.tick_params(labelsize=20)
plt.plot(Tts/Ts, min_lam, linewidth=3)
# plt.title("Lambda vs. Tt", fontsize=20)
plt.xlabel(r'T$_\mathrm{t}$/T$_\mathrm{s}$', fontsize=20)
plt.ylabel(r'Optimal $\lambda$', fontsize=20);

f1.tight_layout()
f1.savefig('Optimal_lambda.eps', bbox_inches='tight')
```
